```python
data = [[],[],[]]
plt.ioff()
with open('roc_all') as infile:
    lines = infile.readlines()
    counter = -1
    for line in lines: 
        for n in line.split():
            if n[:8] == '(array([':
                counter =0
                data[counter].append(float(n[8:-1]))
            elif n[:7] == 'array([':
                counter +=1
                if counter < 2: 
                    data[counter].append(float(n[7:-1]))
            elif n[-3:] == "]),":
                if counter < 2:
                    data[counter].append(float(n[:-3]))
            else:
                if counter < 2:
                    data[counter].append(float(n[:-1]))

                        
plt.figure()
plt.plot(data[0], data[1])
plt.plot([0,1],[0,1], "--", color = "black")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend(frameon=False, title = r"Variant-All $a_1^{\pm} - \rho^{\mp}$")
plt.savefig("roc_all.eps")
plt.close()
```
